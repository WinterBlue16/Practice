```python
# 200123 keras DNN 강의 practice

# 1. 라이브러리 불러오기
import numpy as np 


# 정제된 데이터 만들기
x = np.array([range(1, 201), range(201, 401), range(401, 601), range(601, 801)])
y = np.array([range(1, 201), range(201, 401)])

# 데이터 형태 확인
print(x.shape)
print(y.shape)
```

```python
# 2. 행, 열 변환하기

x = np.array([range(1, 201), range(201, 401), range(401, 601), range(601, 801)])
y = np.array([range(1, 201), range(201, 401)]) # 왜 갯수가 50개여야 하나?

x = np.transpose(x)
y = np.transpose(y)

# 3. 데이터 train, test, validation으로 나누기
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4)
x_val,x_test, y_val, y_test = train_test_split(x_test, y_test, test_size=0.5)

# 4. 모델 만들기

from keras.models import Sequential
from keras.layers import Dense, Dropout

model = Sequential()
model.add(Dense(256, input_dim = 4))
model.add(Dense(256))
model.add(Dense(256))
model.add(Dense(2))

# 5. complie & 모델 학습시키기

model.compile(loss = 'mse', optimizer='adam',
             metrics = ['mse'])
model.fit(x_train, y_train, epochs=50, batch_size=1, validation_data=(x_val, y_val))

# 6. 평가 예측하기 

loss, mse = model.evaluate(x_test, y_test, batch_size=1)
print('mse :', mse)

# 7. 새로운 데이터 넣어보기

# x_prd = np.array([[254, 655, 245], [865, 348, 347], [877, 347, 289], [327, 854, 893]]).T
y_prd = model.predict(x_test)
# print(y_prd)

# 8. RMSE 구하기
from sklearn.metrics import mean_squared_error

def RMSE(y_test, y_prd):
    return np.sqrt(mean_squared_error(y_test, y_prd))
print('RMSE :', RMSE(y_test, y_prd))

# 9. R2 구하기

from sklearn.metrics import r2_score

r2_y_prd = r2_score(y_test, y_prd)
print('R2 :', r2_y_prd)
```

```python
# 200127 DNN Practice

import numpy as np 

x = np.array([range(1, 51), range(51, 101), range(101, 151), range(151, 201)])
y = np.array([range(51, 101)])

x = x.T
y = y.T

from sklearn.model_selection import train_test_split 

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4)
x_test, x_val, y_test, y_val = train_test_split(x_test, y_test, test_size=0.5)

from keras.models import Sequential
from keras.layers import Dense


model = Sequential()

model.add(Dense(128, input_shape=(4,)))
model.add(Dense(64))
model.add(Dense(32))
model.add(Dense(1))

model.compile(loss='mae', optimizer='adam',
             metrics=['mae'])

model.fit(x_train, y_train, epochs=100, batch_size=2, validation_data=(x_val, y_val))

loss, mae = model.evaluate(x_test, y_test, batch_size=2)
print('mae :', mae)

# x_prd = np.array([[753,234,125],[435,768,326], [235,895,294], [928, 464, 754]])
# x_prd = np.transpose(x_prd)
result = model.predict(x_test)
print(result)

from sklearn.metrics import mean_absolute_error

def RMAE(y_test, result):
    return np.sqrt(mean_absolute_error(y_test, result))
print('RMAE :', RMAE(y_test, result))


from sklearn.metrics import r2_score

model_r2 = r2_score(y_test, result)
print('R2 :', model_r2)
    
```
