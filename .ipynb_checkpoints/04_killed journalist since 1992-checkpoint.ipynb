# 2. 데이터 전처리하기
- 이제 본격적인 데이터 전처리에 들어간다. 우선 모든 str을 숫자 형태로 바꿔주어야 한다. 

```python
df.Date.value_counts()
```

```python
# 1992년에 해당하는 값들을 모두 1992로 통일 
df1 = df[df['Date'].str.contains('1992')==True]
df1['Date'] = 1992
df1 # 1992만 포함된 df
```

```python
# 반복작업을 줄이기 위한 함수 생성

# def changedate(df, column, num):
#     if column.str.contains(str(num))==True:
#         x = df[column.str.contains(str(num))==True]
#         x = num
#         return x
#     else:
#         pass

def changedate(df, num):
    if df['Date'].str.contains(str(num))==True:
        x = df[df['Date'].str.contains(str(num))==True]
        x = num
        return x # 오류 발생
```

```python
changedate(df, 1993)
```
