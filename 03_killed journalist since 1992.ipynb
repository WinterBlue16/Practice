```python
# 1. 많이 죽은 나라순으로 묶고 나머지 기타 처리하기 => 2. 소속 대륙별로 묶고 분류하기
df['Country_killed'] = df['Country_killed'].replace(['Thailand', 'Pakistan', 'Afghanistan', 'India', 'Bangladesh', 
                                                     'Myanmar', 'Palestine', 'Philippines', 'Indonesia', 'Sri Lanka', 
                                                     'Nepal', 'Kazakhstan', 'Uzbekistan', 'Vietnam', 'Japan',  'Armenia', 'Turkmenistan'], 'Asia') # 아시아
df['Country_killed'] = df['Country_killed'].replace(['Ukraine', 'Poland', 'France', 'Russia', 'Bosnia', 
                                                     'Ireland', 'Spain', 'Greece', 'Lithuania', 'Latvia'], 'Europe') # 유럽
df['Country_killed'] = df['Country_killed'].replace(['Iraq', 'Syria', 'Yemen', 'Turkey', 'Lebanon', 
                                                     'Bahrain', 'Kuwait', 'Saudi Arabia', 'Cyprus'], 'Middle-East') # 중동
df['Country_killed'] = df['Country_killed'].replace(['Sierra Leone', 'Rwanda', 'Algeria', 'Libya', 'Guinea', 
                                                     'Somalia', 'Ghana', 'Democratic Republic of Congo', 'South Sudan', 
                                                     'Central African Republic', 'Egypt', 'Mali', 'Tanzania', 'Nigeria', 
                                                     'Gabon', 'Cameroon', 'Gambia', 'Tunisia', 'Burkina Faso', 'Zimbabwe'], 'Africa') # 아프리카
df['Country_killed'] = df['Country_killed'].replace(['Brazil', 'Mexico', 'Columbia', 'America', 'Guatemala', 
                                                     'Paraguay', 'Honduras', 'Peru', 'Canada', 'Costa Rica', 'Panama'], 'America') # 아메리카
df['Country_killed'] = df['Country_killed'].replace('Papua New Guinea', 'Oceaina') # 오세아니아

df['Country_killed'].value_counts()
```

```python
# 1992년에 해당하는 값들을 모두 1992로 통일 
df1 = df[df['Date'].str.contains('1992')==True]
df1['Date'] = 1992
df1 # 1992만 포함된 df
```

```python
# 같은 연도 행들만 모아주기 위한 함수
def changedate(df, num):
    temp_df = df[df['Date'].str.contains(str(num))==True]
    return temp_df
```

```python
# 전처리 후 데이터프레임 병합 준비
temp_df = changedate(df, 1992)
temp_df['Date'] = 1992
temp_df
```

```python
# 같은 연도수인 행들을 연 단위로 통일
for i in range(1993, 2017):
    sub_df = changedate(df, i) # 같은 연도수인 행만 추출해 데이터프레임화
    sub_df['Date']=i # 연일월을 연으로 일원화
    temp_df = pd.concat([temp_df, sub_df]) # 이미 전처리해 둔 열과 병합 # 중괄호 빼먹지 말 것

temp_df
```

```python
temp_df.rename(columns = {"Date": "Year"}, inplace = True)
temp_df # 열 이름 바꿔주기(Date => Year)
```
